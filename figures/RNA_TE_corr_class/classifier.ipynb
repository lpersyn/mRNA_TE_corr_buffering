```python
from sklearn.calibration import CalibrationDisplay
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import PrecisionRecallDisplay
import pandas as pd
import matplotlib.pyplot as plt

def plots(preds):
    disp = CalibrationDisplay.from_predictions(
        preds["bio_source_buffering_category_true"],
        preds["bio_source_buffering_category_pred"],
        n_bins=10,
        name="LGBM",
    )
    plt.show()
    cm = ConfusionMatrixDisplay.from_predictions(
        preds["bio_source_buffering_category_true"],
        preds["bio_source_buffering_category_pred"].round(),
    )
    plt.title("Confusion Matrix")
    plt.show()
    roc = RocCurveDisplay.from_predictions(
        preds["bio_source_buffering_category_true"],
        preds["bio_source_buffering_category_pred"],
        plot_chance_level = True,
    )
    plt.title("ROC Curve")
    plt.show()
    prc = PrecisionRecallDisplay.from_predictions(
        preds["bio_source_buffering_category_true"],
        preds["bio_source_buffering_category_pred"],
        plot_chance_level = True
    )

    plt.title("Precision-Recall Curve")
    plt.show()
```

```python
preds = pd.read_csv("../../results/human/RNA_TE_corr_classification/lgbm_classification-LL_P5_P3_CF_AAF_3mer_freq_5/predictions.csv")
plots(preds)
```

```python
preds = pd.read_csv("../../results/human/RNA_TE_corr_classification/lgbm_classification-LL_P_P5_PC_P3_WP_K_CF_AAF_DC_1mer_freq_2mer_freq_3mer_freq_4mer_freq_5mer_freq_6mer_freq_Struct/predictions.csv")
plots(preds)
```
